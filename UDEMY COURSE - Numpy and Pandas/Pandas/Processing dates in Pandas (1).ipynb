```python
import pandas as pd
```

```python
data_1 = pd.DataFrame({"Group":["a","a","a","b","b","b","c","c","c",],"Data":[4,2,9,89,66,33,5,7,9]})
data_1
```

```python
data_1.sort_values(by=["Group","Data"], ascending=[False,True], inplace=True)
data_1
```

```python
data_2 = pd.DataFrame({"k1":["one"]*3 + ["two"]*4,"k2":[45,1,5,9,87,3,5]})
data_2
```

```python
data_2.sort_values(by="k2")
```

```python
data_2.drop_duplicates()
```

```python
data_2.drop_duplicates(subset="k1")
```

```python
data_3 = pd.DataFrame({"food":["A1","A2","B1","B2","B3","C1","C2"],"data":[1,2,3,4,5,6,7]})
data_3
```

```python
food_dict = {
    "A1":"A",
    "A2":"A",
    "B1":"B",
    "B2":"B",
    "B3":"B",
    "C1":"C",
    "C2":"C"
}
data_3["food_1"] = data_3["food"].map(food_dict)
data_3
```
